```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Mall_Customers.csv')
```

```python
x=df.iloc[ : , 3:]
```

```python
#import clustering class of k means
from sklearn.cluster import KMeans, AgglomerativeClustering
km=KMeans(n_clusters=3)
km.fit_predict(x)
```

```python
#Find the  best suited value if K ie the number of clusters using elbow method
#sse: sum squared mean We are going to find sse for 1-15 k values and the point at which elbow occurs is the value of k

sse=[]
for k in range(1,16):
    km=KMeans(n_clusters=k)
    km.fit_predict(x)
    sse.append(km.inertia_)

```

```python
sse
```

```python
#plot to find the elbow point
plt.title('Elbow method')
plt.xlabel('Value of K')
plt.ylabel('SSE')
plt.grid()
plt.xticks(range(1,16))
plt.plot(range(1,16),sse,marker='.',color='red')
```

```python
#Elbow point is 5. Therefore Value of K=5
km=KMeans(n_clusters=5)
labels=km.fit_predict(x)
```

```python
labels
```

```python
df.columns
```

```python
#calculate centroids
cent=km.cluster_centers_
```

```python
#plot using kmeans
plt.figure(figsize=(16,9))
plt.title('Clustered Data')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.scatter(x['Annual Income (k$)'],x[ 'Spending Score (1-100)'],c=labels)
plt.scatter(cent[:,0], cent[:,1],s=100,color='k')
```

```python
#Compare unclustered and clustered

plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title('UnClustered Data')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.scatter(x['Annual Income (k$)'],x[ 'Spending Score (1-100)'])

plt.subplot(1,2,2)
plt.title('Clustered Data')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.scatter(x['Annual Income (k$)'],x[ 'Spending Score (1-100)'],c=labels)
plt.scatter(cent[:,0], cent[:,1],s=100,color='k')
```

```python
#using agglomerative Algorithm
agl=AgglomerativeClustering(n_clusters=5)
alabels=agl.fit_predict(x)
alabels
```

```python
plt.figure(figsize=(16,9))
plt.title('Agglomerative Algorithm')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.scatter(x['Annual Income (k$)'],x[ 'Spending Score (1-100)'],c=alabels)
```

```python
#compare 2 clustering algorithms
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title('K-Means')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.scatter(x['Annual Income (k$)'],x[ 'Spending Score (1-100)'],c=labels)
plt.scatter(cent[:,0], cent[:,1],s=100,color='k')

plt.subplot(1,2,2)
plt.title('Agglomerative Algorithm')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.scatter(x['Annual Income (k$)'],x[ 'Spending Score (1-100)'],c=alabels)
```
